```python
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}". It was renamed to own-file.wav'.format(
      name=fn))
  file = open('own-file.wav', 'wb')
  file.write(uploaded[fn])
```
